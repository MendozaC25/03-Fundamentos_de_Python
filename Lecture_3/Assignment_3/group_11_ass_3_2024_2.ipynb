This loop must iterate over **2019, 2018, 2017, 2016, 2015 folders** and import these files. All the files have the same name. You must store these files and their labels in a nested dictionary named as `all_data`. The keys of the dictionary should be named as `year_2019`, for example, and the keys of the nested dictionary should be `data`, `var_labels`, and `value_labels`. **Hint: Use [this link](https://notebooks.githubusercontent.com/view/ipynb?browser=chrome&color_mode=auto&commit=4d6de78e00e7001f16bf6473c2eb7ce24fb611cd&device=unknown_device&enc_url=68747470733a2f2f7261772e67697468756275736572636f6e74656e742e636f6d2f616c6578616e6465727175697370652f4469706c6f6d61646f5f505543502f346436646537386530306537303031663136626636343733633265623763653234666236313163642f4c6563747572655f342f4c6563747572655f342e6970796e62&logged_in=true&nwo=alexanderquispe%2FDiplomado_PUCP&path=Lecture_4%2FLecture_4.ipynb&platform=windows&repository_id=427747212&repository_type=Repository&version=95#4.2.)**

```python
# First, we start with an empty container to store data and labels
all_data = {}
```

```python
#Establishing the years required
years = [2019, 2018, 2017, 2016, 2015]
```

```python
# Base path for the loop
base_path = "../../_data/endes/"
```

```python
# Iterate over each year to import the corresponding files
for year in years:
    try:
        # Building the file paths for each of the three .sav files
        file_names = ["REC0111.sav", "RE223132.sav", "RE516171.sav"]
        
        # Read and process each file for the year
        rec_1, meta_1 = pyreadstat.read_sav(os.path.join(base_path, str(year), file_names[0]))
        rec_2, meta_2 = pyreadstat.read_sav(os.path.join(base_path, str(year), file_names[1]))
        rec_3, meta_3 = pyreadstat.read_sav(os.path.join(base_path, str(year), file_names[2]))
        
        # Selecting the columns for each dataframe
        cols1 = ['CASEID', 'V000', 'V001', 'V002', 'V003', 'V004', 'V008', 'V009', 'V010', 'V011', 'V012', 'V024', 'V102', 'V120', 'V121', 'V122', 'V123', 'V124', 'V125', 'V127', 'V133']
        rec1_1 = rec_1.loc[:, cols1]
        
        cols2 = ['CASEID', 'V201', 'V218', 'V301', 'V302', 'V323', 'V323A', 'V325A', 'V326', 'V327', 'V337', 'V359', 'V360', 'V361', 'V362', 'V363', 'V364', 'V367', 'V372', 'V372A', 'V375A', 'V376', 'V376A', 'V379', 'V380']
        rec2_1 = rec_2.loc[:, cols2]
        
        cols3 = ['CASEID', 'V501', 'V502', 'V503', 'V504', 'V505', 'V506', 'V507', 'V508', 'V509', 'V510', 'V511', 'V512', 'V513', 'V525', 'V613', 'V714', 'V715']
        rec3_1 = rec_3.loc[:, cols3]
        
        # Generate a new column for rec1_1, named "year"
        rec1_1['year'] = year
        
        # Update the variable labels
        var_labels = {}
        var_labels.update(meta_1.column_names_to_labels)
        var_labels.update(meta_2.column_names_to_labels)
        var_labels.update(meta_3.column_names_to_labels)
        
        # Update the value labels 
        value_labels = {}
        value_labels.update(meta_1.variable_value_labels)
        value_labels.update(meta_2.variable_value_labels)
        value_labels.update(meta_3.variable_value_labels)
        
        # Merging the datasets using 'CASEID'
        endes_data = rec1_1.merge(rec2_1, on='CASEID', how='inner')
        endes_data = endes_data.merge(rec3_1, on='CASEID', how='inner')
        
        # Storing the data, var_labels, and value_labels in the loop
        all_data[f'year_{year}'] = {
            'data': endes_data,
            'var_labels': var_labels,
            'value_labels': value_labels
        }
    
    except Exception as e:
        print(f"Error occurred while processing files for year {year}: {e}")
```

```python
# Print the structure of all_data to verify the loop procedure
print("\nAll Data Dictionary Structure:")
for year_key in all_data:
    print(year_key)
    print("  Data Sample:")
    print(all_data[year_key]['data'].head())  # Print 5 lines of the data
    print("  Variable Labels:")
    print(list(all_data[year_key]['var_labels'].items())[:5])  # Display the first 5 variable labels
    print("  Value Labels:")
    print(list(all_data[year_key]['value_labels'].items())[:5])  # Display the first 5 value labels
    print("\n")
```

7. Use `all_data` to append all the data sets. Store all data sets in a list using `for loop`. Then, use `pd.concat` to append all the data sets. This new object should be named as `endes_data_2015_2019`. **Hint: Use [this code](https://stackoverflow.com/questions/32444138/concatenate-a-list-of-pandas-dataframes-together)**

Use `endes_data_2015_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**
