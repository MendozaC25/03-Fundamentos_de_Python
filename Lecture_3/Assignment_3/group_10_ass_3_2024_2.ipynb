1. Import the `REC0111.sav`, `RE223132.sav` and `RE516171.sav` files and their variables and values labels from this path `"../../_data/endes/2019"`.
```python
##1

# Define file paths and names

file_paths = [
    "../../_data/endes/2019/REC0111.sav",
    "../../_data/endes/2019/RE223132.sav",
    "../../_data/endes/2019/RE516171.sav"]

# Names of files
names = ["rec_1", "rec_2", "rec_3"]

# Initialize dictionaries
data_frames = {}
var_labels = {}
value_labels = {}

# Function to load .sav files

def load_sav_file(file_path):
    df, meta = pyreadstat.read_sav(file_path)
    return df, dict(meta.column_names_to_labels), dict(meta.variable_value_labels)

# Load files and update dictionaries
for name, path in zip(names, file_paths):
    df, var_label, value_label = load_sav_file(path)
    data_frames[name] = df
    var_labels[name] = var_label
    value_labels[name] = value_label

# Access specific data frames and labels
rec_1 = data_frames["rec_1"]
rec_2 = data_frames["rec_2"]
rec_3 = data_frames["rec_3"]

var_labels1 = var_labels["rec_1"]
value_labels1 = value_labels["rec_1"]

var_labels2 = var_labels["rec_2"]
value_labels2 = value_labels["rec_2"]

var_labels3 = var_labels["rec_3"]
value_labels3 = value_labels["rec_3"]

# Print variable and value labels for rec_1
print(value_labels1)
print(var_labels1)
```

7. Use `all_data` to append all the data sets. Store all data sets in a list using `for loop`. Then, use `pd.concat` to append all the data sets. This new object should be named as `endes_data_2015_2019`. **Hint: Use [this code](https://stackoverflow.com/questions/32444138/concatenate-a-list-of-pandas-dataframes-together)**

```python
##7

# Initialize an empty list to store all dataframes
data_frames = []

# Loop through each year in the all_data dictionary
for year_key in all_data.keys():
    # We have to append the df for the current year to the list
    # We use append because we're adding individual df, not lists of df (which is what .extend does)
    data_frames.append(all_data[year_key]['data'])

# Lastly, we concatenate all df into a single df row-wise (axis=0)
 # Why row-wise? Because our df have the same columns but represent different cases (persons)
 # We also set 'ignore_index = True' to avoid having repeated indexes (0, 1, 2, 4, 0, 1, 2, 3...)
endes_data_2015_2019 = pd.concat(data_frames, axis=0, ignore_index=True)

# Checking the result
print(endes_data_2015_2019)
```
