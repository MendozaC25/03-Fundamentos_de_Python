### Pregunta 6

Get the min, max, sd, n_obs, n_missing for the following columns total children ever born (V201), Ideal number of children (V613), Husbands education-single yrs (V715), and Age at first marriage (V511). We want a dataframe with the following columns Variables, Min, Max, Mean, N_obs, N_missing and sort by the number of missing rows. Hint: Use describe and pivot methods

```python
# obtengo toda la estadistica descriptiva y hago la transpuesta
data_selected = endes_2019.loc[ :, ["V201" ,"V613","V715","V511"]]
#incluyo todos los datos hasta los na para obtener las estadisticas
descriptive_stats=data_selected.describe(include = 'all').reset_index()
descriptive_stats
```

```python
## verifico los tipos de datos
descriptive_stats.dtypes
```

```python
## verifico la forma de los datos
print(f"los datos unicos de V201 {data_selected['V201'].unique()}")
print(f"los datos unicos de V613 {data_selected['V613'].unique()}")

print(f"los datos unicos de V715 {data_selected['V715'].unique()}")

print(f"los datos unicos de V511 {data_selected['V511'].unique()}")
```

```python
# renombro las columnas como me las piden y cambio el nombre al indice para trabajar
descriptive_stats.rename(columns={"index": "stats", "count": "n_obs"},inplace=True)
descriptive_stats
```

```python
# acomodo los datos para la funcion pivot con la funcion melt
descriptive_stats = descriptive_stats.melt(id_vars = ['stats'], var_name = 'Variables', value_name = 'values')
descriptive_stats
```

```python
#aplico la pivot para obtener los datos, creo los n_missing
descriptive_stats = descriptive_stats.pivot(index = 'Variables', columns = 'stats', values = 'values').reset_index()
descriptive_stats.loc[:,'n_missing'] = list(data_selected.loc[:,['V201','V613','V715','V511']].isnull().sum())
descriptive_stats
```

```python
#renombro variables, seleecciono y ordeno variables
descriptive_stats.rename(columns={"count": "N_obs","min": "Min", "max": "Max","mean": "Mean","n_missing": "N_missing"},inplace=True)
descriptive_stats = descriptive_stats.loc[:,['Variables', 'Min', 'Max', 'Mean', 'N_obs', 'N_missing']]
descriptive_stats = descriptive_stats.sort_values('N_missing')
descriptive_stats
```

### Pregunta 7

Use endes_2019 data to generate a new object named mean_key_vars to find the mean of total children ever born (V201), Ideal number of children (V613), Husbands education-single yrs (V715), and Age at first marriage (V511) by year and department (V024). Name these columns as mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage, respectively. Hint: Use groupby and this link.

```python
#selecciono los datos y verifico tipo de datos

datos = endes_2019.loc[:,[ "V201", "V613", "V715", "V511","year","V024" ]]
datos.dtypes
```

```python
## transformo a numero y verifico tipo
datos[["V715", "V613"]] = datos[['V613','V715']].apply(pd.to_numeric, errors = 'coerce' )
datos.dtypes
```

```python
## agrupo y saco la media
mean_key_vars = datos.groupby(['year', 'V024']).mean()
mean_key_vars
```

```python
## cambio los nombres de las columnas
mean_key_vars = mean_key_vars.rename(columns = {"V201": "Mean_total_children",
                          "V613": "Ideal number of children",
                          "V715": "Husbands education-single yrs",
                          "V511": "Age at first marriage"})
mean_key_vars
```

### Pregunta 8
Reshape mean_key_vars from wide to long. We want a dataframe with three columns dpto, variables, values. Name this object as reshape_mean_key_vars. Hint: Use melt method.

```python
#elimino los indices para trabajar
mean_key_vars.reset_index(inplace=True)
```

```python
# creo el nuevo objeto
reshape_mean_key_vars = mean_key_vars.melt(id_vars = ['V024'], var_name = 'Variables', value_name = 'values')
# cambio de nombre a la variable
reshape_mean_key_vars.rename(columns={"V024" : "dpto"},inplace=True)
reshape_mean_key_vars
```
