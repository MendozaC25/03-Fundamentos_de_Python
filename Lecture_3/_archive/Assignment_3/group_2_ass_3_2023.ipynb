7. Use endes_2019 data to generate a new object named mean_key_vars to find the mean of total children ever born (V201), Ideal number of children (V613), Husbands education-single yrs (V715), and Age at first marriage (V511) by year and department (V024). Name these columns as mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage, respectively. Hint: Use groupby and this link.

First: We group the data by the columns 'V024' and 'year' using groupby. Then, it calculates the mean for the specified columns ('V201', 'V613', 'V715', 'V511') for each group using the agg function. The result is stored in a new object called mean_key_vars. The reset_index() function is used to reset the index of the resulting DataFrame.

```python
mean_key_vars = endes_2019.groupby(['V024', 'year']).agg({
    'V201': 'mean',                 
    'V613': 'mean',                 
    'V715': 'mean',                 
    'V511': 'mean'                  
}).reset_index()
```

Then, we rename the columns of the mean_key_vars object to the names provided.

```python
mean_key_vars.columns = ['V024', 'year', 'mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage']
```

```python
print(mean_key_vars)
```

8. Reshape mean_key_vars from wide to long. We want a dataframe with three columns dpto, variables, values. Name this object as reshape_mean_key_vars. Hint: Use melt method.

```python
reshape_mean_key_vars = pd.melt(mean_key_vars, id_vars=['V024', 'year'], value_vars=['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage'])
```

We reshape the objecto using melt to reshape from wide to long format. The id_vars parameter specifies the columns to be retained as identifier variables ('V024' and 'year' in this case), and value_vars specifies the columns to be melted.

Then, we rename the colums of the columns melted.

```python
reshape_mean_key_vars.columns = ['dpto', 'year', 'variables', 'values']
```

```python
print(reshape_mean_key_vars)
```

9. Replicate your answers for questions 7 and 8, but in one line of code. Make it the most simple as possible. NO HINT

We use the chaining method to chain all the operations together into one single line of code. So just like questions 7 and 8, we group the data by the columns 'V024' and 'year' . Then, it calculates the mean for the specified columns ('V201', 'V613', 'V715', 'V511') and rename the columns according to to the names provided.  And finally, it reshapes the object from long to wide using 'V024' and 'year as ID parameters. (We skip changing the names of the final columns to be able to merge it with endes_2019 later). So with the chaining method all the DataFrame manipulation methods are sequentially applied in one statement. 

```python
reshape_mean_key_vars_2 = pd.melt(
    endes_2019.groupby(['V024', 'year']).agg({
        'V201': 'mean',
        'V613': 'mean',
        'V715': 'mean',
        'V511': 'mean'
    }).reset_index().rename(columns={
        'V201': 'mean_total_children',
        'V613': 'mean_ideal_children',
        'V715': 'mean_hb_yr_educ',
        'V511': 'mean_first_marriage'
    }), 
    id_vars=['V024', 'year'],
    value_vars=['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage']
)
```

```python
print(reshape_mean_key_vars_2)
```

10. Merge reshape_mean_key_vars with endes_2019. Name this object final_result. Hint: Use merge.

We merge mean_key_vars with endes_2019 with the panda funciton 'merge' using 'V024', 'year' as common keys on which the merge operation will be based. 

```python
final_result = pd.merge(endes_2019, reshape_mean_key_vars_2, on=['V024', 'year'])
```
