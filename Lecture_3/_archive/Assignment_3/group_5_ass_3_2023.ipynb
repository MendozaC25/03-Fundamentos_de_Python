6. Get the min, max, sd, n_obs, n_missing for the following columns **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)**. We want a dataframe with the following columns **Variables, Min, Max, Mean, N_obs, N_missing** and sort by the number of missing rows. **Hint: Use `describe` and `pivot` methods**

```python
# solution:
# =========
# Select specified columns
selected_columns = ['V201', 'V613', 'V715', 'V511']

# Get descriptive statistics using "describe"
descriptive_stats = endes_2019[selected_columns].describe()

# Calculate the number of non-null rows and the number of missing rows for each column
n_obs = endes_2019[selected_columns].count()
n_missing = endes_2019.shape[0] - n_obs

# Create a DataFrame with the number of missing items and observations
missing_obs_df = pd.DataFrame({
    'Variables': selected_columns,
    'N_obs': n_obs,
    'N_missing': n_missing
})

# Use the pivot method to organize descriptive statistics
result_df = descriptive_stats.T.reset_index()

# Merge with the DataFrame of missing and observations
result_df = result_df.merge(missing_obs_df, left_on='index', right_on='Variables')

# Rearrange the DataFrame to have the requested columns
result_df = result_df[['Variables', 'min', 'max', 'mean', 'N_obs', 'N_missing']]

# Rename the columns
result_df.columns = ['Variables', 'Min', 'Max', 'Mean', 'N_obs', 'N_missing']

# Sort the DataFrame by the number of missing items
result_df = result_df.sort_values(by='N_missing', ascending=False)

# print the resulting DataFrame
result_df.head()
```

```python
# Use stack and unstack to organize descriptive statistics in a different way
result_pivot = result_df.set_index('Variables').stack().unstack(level=0)

# print the resulting DataFrame
result_pivot.head()
```

7. Use `endes_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

```python
# We print the requested variables to be analyzed to see their presentation in the data frame:
print(endes_2019[['year', 'V024', 'V201', 'V613', 'V715', 'V511']])
```

```python
# We analyze the type of variable to check that they are numeric:
selected = endes_2019[['year', 'V024', 'V201', 'V613', 'V715', 'V511']].dtypes
print(selected)
```

```python
# We group each requested variable by year and department code:
mean_total_children = endes_2019.groupby(['year', 'V024'])['V201'].mean().reset_index()
mean_ideal_children = endes_2019.groupby(['year', 'V024'])['V613'].mean().reset_index()
mean_hb_yr_educ = endes_2019.groupby(['year', 'V024'])['V715'].mean().reset_index()
mean_first_marriage = endes_2019.groupby(['year', 'V024'])['V511'].mean().reset_index()
```

```python
# We create the new dataframe from the groups of variables using the merge function
mean_key_vars = mean_total_children.merge( mean_ideal_children, on = [ 'year' ,'V024' ] , how = "left" , validate = "m:1" )
mean_key_vars = mean_key_vars.merge( mean_hb_yr_educ, on = [ 'year' ,'V024' ] , how = "left" , validate = "m:1" )
mean_key_vars = mean_key_vars.merge( mean_first_marriage, on = [ 'year' ,'V024' ] , how = "left" , validate = "m:1" )

# We rename the columns with the name of each grouped variable
mean_key_vars = mean_key_vars.rename(columns={'V024': 'dpto', 'V201': 'mean_total_children', 'V613': 'mean_ideal_children', 'V715': 'mean_hb_yr_educ', 'V511': 'mean_first_marriage'})

# We show the first variables of the new dataframe
mean_key_vars.head()
```

8. Reshape `mean_key_vars` from wide to long. We want a dataframe with three columns **dpto, variables, values**. Name this object as `reshape_mean_key_vars`. **Hint: Use melt method**.

```python
# We reshape the dataframe from wide to long using the melt method, considering only the variables generated (excluding year)
reshape_mean_key_vars = mean_key_vars.melt( 
                    id_vars = [ 'dpto' ] ,
                    value_vars=['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage'],
                    var_name = 'variables', 
                    value_name = 'values'
                 )
```

```python
# Sort the new dataframe by department code:
reshape_mean_key_vars
reshape_mean_key_vars.sort_values(by=['dpto'])
```

9. Replicate your answers for questions 7 and 8, but in one line of code. Make it the most simple as possible. **NO HINT**

```python
# Combining both operations (questions 7 and 8) into a single line
reshape_mean_key_vars = endes_2019.groupby(['V024', 'year']).agg(
    dpto=('V024', 'first'),
    mean_total_children=('V201', 'mean'),
    mean_ideal_children=('V613', 'mean'),
    mean_hb_yr_educ=('V715', 'mean'),
    mean_first_marriage=('V511', 'mean')
).reset_index().melt(id_vars=['dpto'], var_name='variables', value_name='values', value_vars=['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage']).sort_values(by=['dpto'])

# We show the first variables of the new dataframe
reshape_mean_key_vars.head()
```

10. Merge `reshape_mean_key_vars` with `endes_2019`. Name this object `final_result`. **Hint: Use merge.**

```python
# Merging both df in one object call "final_result"
final_result = pd.merge(endes_2019, reshape_mean_key_vars, left_on=['V024'], right_on=['dpto'])

# We show the first variables of the new dataframe
final_result.head()
```
