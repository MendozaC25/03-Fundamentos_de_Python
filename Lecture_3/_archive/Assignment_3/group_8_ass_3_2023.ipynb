## 6. Get the min, max, sd, n_obs, n_missing for the following columns **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)**. 
We want a dataframe with the following columns **Variables, Min, Max, Mean, N_obs, N_missing** and sort by the number of missing rows. **Hint: Use `describe` and `pivot` methods**

```python
df= endes_2019[["V201","V715","V613","V511"]]
#Create a variable with the description of the variables
df1 = df.describe(include = 'all')
#Add the news indexes
df1.loc['n_obs'] = len(df)  
df1.loc['n_missing'] = df.isnull().sum()
df1
```

```python
#Select the required indexes and change the name of the index to "variables"
df2 = df1.loc[["min", "max", "std", "mean", "count", "n_missing"]].reset_index().rename(columns = {'index':'Variables'})
#Pivot the dataframe and rename the columns. Then, we sort the data according to number of missing values. 
df2 = df2.pivot_table(columns= 'Variables').rename(columns=str.capitalize).sort_values(by= 'N_missing').rename(columns = {'Count':'N_obs'})
df2 = df2[["Min","Max","Mean","N_obs","N_missing"]]
df2
```

## 7. Use `endes_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. 
Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

```python
#convert into numeric variables the categorical ones using the method shown in the link of the exercise
endes_2019.V715 = pd.Categorical(endes_2019.V715)
endes_2019['V715_code'] = endes_2019.V715.cat.codes

endes_2019.V613 = pd.Categorical(endes_2019.V613)  
endes_2019['V613_code'] = endes_2019.V613.cat.codes
```

```python
#create a new object (mean_key_vars) using groupby to agrupate dataframe by year and department (V024). Then  we specify which variables will be used  as required on the exercise.
#To find the mean of those variables we use the function mean and rename each variable by the ones required in the correct order of columns.

mean_key_vars = endes_2019.groupby(["V024"], as_index=False)[["V201", "V613_code", "V715_code", "V511"]].mean()
mean_key_vars.rename(columns={"V201": "mean_total_children",  "V613_code": "mean_ideal_children",  "V715_code": "mean_hb_yr_educ", "V511":"mean_first_marriage"}, inplace=True)
mean_key_vars
```

## 8. Reshape `mean_key_vars` from wide to long. 
We want a dataframe with three columns **dpto, variables, values**. Name this object as `reshape_mean_key_vars`. **Hint: Use melt method**.

```python
#Using melt mehod to massage the dataframe into a format where one column is an identifier variable (id_vars),
#while all other columns are measured variables. 
#Then we indicate the new name of the three columns
mean_key_vars = mean_key_vars.reset_index(inplace=False)
reshape_mean_key_vars = mean_key_vars.melt(id_vars = "V024", value_vars = ["mean_total_children", "mean_ideal_children", "mean_hb_yr_educ", "mean_first_marriage"])
reshape_mean_key_vars.columns = ['dpto', 'variables', "values"]
reshape_mean_key_vars
```
