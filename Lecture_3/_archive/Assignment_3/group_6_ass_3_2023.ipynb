Get the min, max, sd, n_obs, n_missing for the following columns **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)**. We want a dataframe with the following columns **Variables, Min, Max, Mean, N_obs, N_missing** and sort by the number of missing rows. **Hint: Use `describe` and `pivot` methods**

```python
# We are selecting the columns of interest:
dataframe = endes_2019[["V201","V715","V613","V511"]]

# Then, we are going to create a variable with the description of the variables:
dataframe1 = dataframe.describe(include = 'all')

#Finally, we are adding the indexes "n_obs" and "n_missing":
dataframe1.loc['n_obs'] = len(dataframe)  
dataframe1.loc['n_missing'] = dataframe.isnull().sum()
```

```python
dataframe1
```

We see that the dataframe has all the indexes as rows, so we have to pivot them and only select: Min, Max, Mean, N_obs, N_missing.

```python
# We are going to select the required indexes.
# and change the name of the selected index to "Variables" as requested:
dataframe2 = dataframe1.loc[["min", "max", "std", "mean", "count", "n_missing"]].reset_index().rename(columns = {'index':'Variables'})

# Now, we are going to pivot the dataframe 
# Rename the columns: adding capital letters, changing the column count to "N_obs".
# Also, we sort the dataframe according to number of missing rows (N_missing). 
dataframe2 = dataframe2.pivot_table(columns= 'Variables').rename(columns=str.capitalize).sort_values(by= 'N_missing').rename(columns = {'Count':'N_obs'})

dataframe2
```

```python
# Finally, we modifify the order of the columns as follows:
dataframe2 = dataframe2[["Min","Max","Mean","N_obs","N_missing"]]
dataframe2
```

7. Use `endes_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

```python
#First, we need to explore the nature of the chosen variables 
var = endes_2019.loc[:,[ "V201", "V613", "V715", "V511","year","V024" ]]
var.dtypes
```

```python
#Second, we transform a numeric the variables that will work with
var[["V715", "V613"]] = var[['V613','V715']].apply(pd.to_numeric, errors = 'coerce' )
```

```python
# In this problem is request to using "groupby" to sort by year and department
mean_key_vars = var.groupby( ['year', 'V024'] )[['V201', 'V613', 'V715','V511']].mean().rename(columns = {'V201' : "mean_total_children" , 'V613': "mean_ideal_children" , 
                                                                                                                   'V715': "mean_hb_yr_educ", 'V511': "mean_first_marriage"}).reset_index()

# We call "mean_key_vars" to see our result 
mean_key_vars.head()
```

8. Reshape `mean_key_vars` from wide to long. We want a dataframe with three columns **dpto, variables, values**. Name this object as `reshape_mean_key_vars`. **Hint: Use melt method**.

```python
#In this case we will use "melt method" to generate a dataframe with the three columns requested
reshape_mean_key_vars = pd.melt(mean_key_vars, id_vars = ['V024'] , value_vars= ["mean_total_children", "mean_ideal_children", "mean_hb_yr_educ" , "mean_first_marriage"], 
                               ).rename(columns = {'V024': 'dpto'})

reshape_mean_key_vars
```
