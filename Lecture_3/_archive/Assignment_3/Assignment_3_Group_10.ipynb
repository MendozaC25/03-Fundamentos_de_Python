7. Use `endes_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

```python
# Select the variables
endes_2019_new = endes_2019[['V024', 'V201', 'V613', 'V715', 'V511']]

# Rename the columns
endes_2019_new.columns = ['department', 'total_children', 'ideal_children', 'hb_yr_educ', 'first_marriage']
endes_2019_new
```

We use the groupby function to summarize the values (obtain means).

```python
# Convert variables to numerical
cols = ['total_children', 'ideal_children', 'hb_yr_educ', 'first_marriage']
endes_2019_new[cols] = endes_2019_new[cols].apply(pd.to_numeric, errors='coerce')

# Calculate averages by year and department
mean_key_vars = endes_2019_new.groupby(['department']).mean()
mean_key_vars
# Rename columns
mean_key_vars.columns = ['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage']
mean_key_vars
```

8. Reshape `mean_key_vars` from wide to long. We want a dataframe with three columns **dpto, variables, values**. Name this object as `reshape_mean_key_vars`. **Hint: Use melt method**.

```python
mean_key_vars['department'] = mean_key_vars.index
mean_key_vars
```

```python
reshape_mean_key_vars = pd.melt(mean_key_vars, id_vars=['department'], value_vars=['mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage'])
reshape_mean_key_vars
```
