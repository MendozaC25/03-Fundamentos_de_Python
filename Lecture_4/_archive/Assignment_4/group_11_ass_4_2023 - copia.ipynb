6. Now, replicate your code of the prevoius sections but for years 2019, 2018, 2017, 2016, 2015. Import the REC0111.sav, RE223132.sav and RE516171.sav files and their variables and values labels from this path "../../_data/endes/". For this excersie you must use a for loop. This loop must iterate over 2019, 2018, 2017, 2016, 2015 folders and import these files. All the files have the same name. You must store these files and their labels in a nested dictionary named as all_data. The keys of the dictionary should be named as year_2019, for example, and the keys of the nested dictionary should be data, var_labels, and value_labels. Hint: Use this link


```python
#  Defining the year we are going to use 
years = [2019, 2018, 2017, 2016, 2015]

# New dictionary to store all data
all_data = {}

# Loop to iterate over the defined years 
for year in years:
    folder_path = f"../../_data/endes/{year}/" #path
    year_dict = {"data": None, "var_labels": None, "value_labels": None}
    
    try:
        #REC0111.sav
        rec1, meta1 = pyreadstat.read_sav(os.path.join(folder_path, "REC0111.sav"))
        year_dict["data"] = rec1
        year_dict["var_labels"] = meta1.variable_value_labels
        year_dict["value_labels"] = meta1.value_labels
        
        #RE223132.sav
        rec2, meta2 = pyreadstat.read_sav(os.path.join(folder_path, "RE223132.sav"))
        year_dict["data"] = rec2
        year_dict["var_labels"].update(meta2.variable_value_labels)
        year_dict["value_labels"].update(meta2.value_labels)
        
        #RE516171.sav
        rec3, meta3 = pyreadstat.read_sav(os.path.join(folder_path, "RE516171.sav"))
        year_dict["data"] = rec3
        year_dict["var_labels"].update(meta3.variable_value_labels)
        year_dict["value_labels"].update(meta3.value_labels)
        
        print(f"Data for {year} loaded")
    except pyreadstat.PyreadstatError as e:
        print(f"ERROR {year}: {e}")

    all_data[f"year_{year}"] = year_dict # Store the new dictionary in the main one

# See the new datasets 
for key, value in all_data.items():
    print(f"\n{key}:")
    print(f"Number of rows: {len(value['data'])}")
    print(f"Variable labels: {value['var_labels']}")
    print(f"Value labels: {value['value_labels']}")
```

7. Use all_data to append all the data sets. Store all data sets in a list using for loop. Then, use pd.concat to append all the data sets. Also, you must reset the index to have a good-looking data. This new object should be named as endes_data_2015_2019. Hint: Use this code

```python
#New empty list 
data_frames_list = []

#Loop to store all data sets
for key, value in all_data.items():
    data_frames_list.append(value['data'])

# Concatenate into a single dataframe
endes_data_2015_2019 = pd.concat(data_frames_list, ignore_index=True)

# Verificate the new data frame 
print(endes_data_2015_2019.head())
```

8. Store all the var_labels and value_labels in a dictionary named as all_var_labels and all_value_labels. The first keys should be the year for both dictionaries.Then, use them to generate new attributes for endes_data_2015_2019. These attributes should be named as var_labels and value_labels.

```python
#Create dictionaries to store labels and variable values.
all_var_labels = {}
all_value_labels = {}

# Iterate over the data for each year and store tags in the dictionaries using "in".
for key, value in all_data.items():
    year = int(key.split('_')[1])  # Extract the year from the key name
    all_var_labels[year] = value['var_labels']
    all_value_labels[year] = value['value_labels']

# Add new attributes to endes_data_2015_2019
endes_data_2015_2019['var_labels'] = endes_data_2015_2019['ID1'].map(all_var_labels)
endes_data_2015_2019['value_labels'] = endes_data_2015_2019['ID1'].map(all_value_labels)

#New data frame whith new attributes
print(endes_data_2015_2019.head())
```

Use `endes_data_2015_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201), Ideal number of children (V613), Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and this link**

```python
# Selecting the variables to mean
selected_columns = ['V201', 'V613', 'V715', 'V511', 'V024', 'year']

# Filter the DataFrame with the selected columns
selected_data = endes_data_2015_2019[selected_columns]

# Rename the columns 
selected_data = selected_data.rename(columns={
    'V201': 'mean_total_children',
    'V613': 'mean_ideal_children',
    'V715': 'mean_hb_yr_educ',
    'V511': 'mean_first_marriage'
})

# Calculate means by year and department
mean_key_vars = selected_data.groupby(['year', 'V024']).mean().reset_index()

# View the new DataFrame
print(mean_key_vars.head())
```
