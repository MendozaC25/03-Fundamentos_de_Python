```python
# We create an empty dictionary to store all the data
all_data = {}

# A list is created with the files to use
files = ["REC0111.sav", "RE223132.sav", "RE516171.sav"]

# A list is created with the variables to use
lists = [list_1, list_2, list_3]

# We iterate through the years 2015 to 2019
for year in range(2015, 2020):
    
    # We create a dictionary for each year
    all_data[f"year_{year}"] = {}
    
    # We create lists that will be useful later
    selected_dfs=[]
    var=[]
    value=[]
    
    # We start the loop
    for i, (file, lists_cell) in enumerate(zip(files, lists), 1):
        
        # We print the name of the file and its corresponding year
        print(f"The loop is processing the file {file} for year {year}")  
        
        # This code allows you to import .sav files
        df, rec = ps.read_sav(f"../../_data/endes/{year}/{file}") 
        
        # We make a loop to search for the columns of the lists in each dataframe
        existing_columns = [col for col in lists_cell if col in df.columns] 
        
        # We print if it finds the columns in the dataframe
        if len(existing_columns) == len(lists_cell):
            print(f"All the variables were found in {file}")
        else:
            for col in lists_cell:
                if col not in existing_columns:
                    print(f"Column {col} were not found in {file}") 
                    
        # Taking this into account, we modify the data for each dataframe with the selected variables
        df_selected = df.loc[:,existing_columns].copy()
        print(f"The filtered data for {file} in year {year} is done")
        
        # In this code we add the labels for the values and variables of the original and filtered data, respectively
        first_var_labels=rec.column_names_to_labels
        first_value_labels=rec.variable_value_labels
        second_var_labels = {var: rec.column_names_to_labels[var] for var in existing_columns if var in rec.column_names_to_labels}
        second_value_labels = {var: rec.variable_value_labels[var] for var in existing_columns if var in rec.variable_value_labels}
        
        # We store the filtered dataframe in the selected_dfs list
        selected_dfs.append(df_selected)

        # We create the year variable only for the filtered dataframe of the first iteration
        if i == 1:
            df_selected.loc[:,"year"]=year
            pass
            
        # We update the label for the year variable only for the first iteration
        if i == 1:
            second_var_labels.update({'year':"Year of the survey"})
            pass    
        
        # We join all the dataframes of each iteration using the merge function
        df_merged = selected_dfs[0]
        for df in selected_dfs[1:]:
            df_merged = df_merged.merge(df, on='CASEID', how='outer', validate = "1:1")
            
        # We update the labels for the values and variables of the dataframe
        var.append(second_var_labels.copy())
        third_var_labels = var[0]
        for y in var[1:]:
            third_var_labels.update(y)
            
        value.append(second_value_labels.copy())
        third_value_labels = value[0]
        for z in value[1:]:
            third_value_labels.update(z)
        
        # We store everything done in the all_data dictionary
        all_data[f"year_{year}"][f"rec{i}"] = df
        all_data[f"year_{year}"][f"rec{i}_{i}"] = df_selected
        all_data[f"year_{year}"][f"var_labels{i}"] = first_var_labels
        all_data[f"year_{year}"][f"value_labels{i}"] = first_value_labels
        all_data[f"year_{year}"][f"new_var_labels{i}"] = second_var_labels
        all_data[f"year_{year}"][f"new_value_labels{i}"] = second_value_labels
        all_data[f"year_{year}"][f"var_labels"] = third_var_labels
        all_data[f"year_{year}"][f"value_labels"] = third_value_labels
        all_data[f"year_{year}"][f"endes_{year}"] = df_merged
        
        print(f"That´s all folks for year {year}")
        
print("That´s all folks")
```

Use `endes_data_2015_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

First, we group the dataFrame by year and department (V024)columns. Using aggs we apply a function to the grouped data. In this case, it calculates the mean for each group for columns 'V201', 'V613', 'V715', and 'V511'. Finally we rename the columns of the resulting DataFrame for better readability.

```python
mean_key_vars = endes_data_2015_2019.groupby(['year', 'V024']).agg({
    'V201': 'mean',
    'V613': 'mean',
    'V715': 'mean',
    'V511': 'mean'
}).reset_index()

# Rename columns
mean_key_vars.columns = ['Year', 'Department', 'mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage']

# Display the resulting DataFrame
print(mean_key_vars)
```
