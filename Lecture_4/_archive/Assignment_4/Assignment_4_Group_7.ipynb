```python
# Defining variables
years = ["2015", "2016", "2017", "2018", "2019"]
files = ["REC0111", "RE223132", "RE516171"]

# To store
all_data = {}
all_var_labels = {} 
all_value_labels = {} 

for year in years:
    all_data_list = []
    var_labels_dict = {}
    value_labels_dict = {}
    
    for file in files:
        data, metadata = pyreadstat.read_sav(f"../../_data/endes/{year}/{file}.sav")
        # Create variable "year"
        anio = int(year) # we use "anio" to dont confuse later
        
        # Keep selected columns
        columns_to_keep = {
            "REC0111": ['CASEID', 'V000', 'V001', 'V002', 'V003', 'V004', 'V008', 'V009', 'V010', 'V011', 'V012', 'V024', 'V102', 'V120', 'V121', 'V122', 'V123', 'V124', 'V125', 'V127', 'V133'],
            "RE223132": ['CASEID', 'V201', 'V218', 'V301', 'V302', 'V323', 'V323A', 'V325A', 'V326', 'V327', 'V337', 'V359', 'V360', 'V361', 'V362', 'V363', 'V364', 'V367', 'V372', 'V372A', 'V375A', 'V376', 'V376A', 'V379', 'V380'],
            "RE516171": ['CASEID', 'V501', 'V502', 'V503', 'V504', 'V505', 'V506', 'V507', 'V508', 'V509', 'V510', 'V511', 'V512', 'V513', 'V525', 'V613', 'V714', 'V715']
        }
        
        selected_columns = columns_to_keep.get(file, [])
        selected_data = data[selected_columns]
        
        # Store information
        year_data = {
            "anio": anio,
            "data_combined": selected_data,
            "var_labels_combined": {col: metadata.column_names_to_labels.get(col, '') for col in selected_columns},
            "value_labels_combined": {col: metadata.variable_value_labels.get(col, {}) for col in selected_columns}
        }
        
        all_data_list.append(year_data)
        
        var_labels_dict.update(metadata.column_names_to_labels)
        value_labels_dict.update(metadata.variable_value_labels)

    all_var_labels[year] = var_labels_dict
    all_value_labels[year] = value_labels_dict
    
    # Get 'all_data'
    all_data[f'year_{year}'] = {
        "anio": anio,
        "data_combined": pd.concat([year_data["data_combined"] for year_data in all_data_list], axis=1),
        "var_labels_combined": var_labels_dict,
        "value_labels_combined": value_labels_dict
    }

# print results
print(all_data)
```

```python
#Use endes_data_2015_2019 data to generate a new object named mean_key_vars
# To find the mean of total children ever born, ideal number of children, husbands education single years 
# and age at first marriage by year and department we use 'groupby()' function and 'agg()' as it's shown 
# in the following lines

mean_key_vars = (
    endes_data_2015_2019
    .loc[:,['anio','V024','V201','V613','V715','V511']] # Keep the variables needed # Consider 'anio' as 'year'
    .groupby( ['anio','V024'], as_index = False ) # Group by 'year' and 'V024' (which contains Departments)
    .agg(Mean_total_children = ( 'V201', np.mean ), # Create new variables as the mean of other
         Mean_ideal_children = ( 'V613', np.mean ),
         Mean_hb_yr_educ = ( 'V715', np.mean ),
         Mean_first_marriage = ( 'V511', np.mean ))
    .round(2) # Round the values at 2 decimals
    .rename(columns = {'anio':'Year'})
    )
mean_key_vars
```
