```python
# File base path
base_path = "../../_data/endes/"

# Years of interest
years = [2019, 2018, 2017, 2016, 2015]

# Dictionary to store all data, variable labels and value labels
all_data = {}

# Loop for each year
for year in years:
    # Build full path for .sav file
    file_path_rec1 = os.path.join(base_path, f"{year}/REC0111.sav")
    file_path_rec2 = os.path.join(base_path, f"{year}/RE223132.sav")
    file_path_rec3 = os.path.join(base_path, f"{year}/RE516171.sav")

    # Read .sav files and get data and metadata
    rec_1, meta_1 = pyreadstat.read_sav(file_path_rec1)
    rec_2, meta_2 = pyreadstat.read_sav(file_path_rec2)
    rec_3, meta_3 = pyreadstat.read_sav(file_path_rec3)

    # Get value and variable labels from sav file
    var_labels1 = meta_1.column_names_to_labels
    value_labels1 = meta_1.variable_value_labels

    var_labels2 = meta_2.column_names_to_labels
    value_labels2 = meta_2.variable_value_labels

    var_labels3 = meta_3.column_names_to_labels
    value_labels3 = meta_3.variable_value_labels

    # Filter required columns in rec1_data, rec2_data and rec3_data
    rec1_data = rec_1.loc[:, ['CASEID', 'V000', 'V001', 'V002', 'V003', 'V004', 
                              'V008', 'V009', 'V010', 'V011', 'V012', 'V024', 'V102', 
                              'V120', 'V121', 'V122', 'V123', 'V124', 'V125', 'V127', 
                              'V133']].copy()
    
    rec2_data = rec_2.loc[:, ['CASEID', 'V201', 'V218', 'V301', 'V302', 'V323', 'V323A', 'V325A',
                              'V326', 'V327', 'V337', 'V359', 'V360', 'V361', 'V362', 'V363',
                              'V364', 'V367', 'V372', 'V372A', 'V375A', 'V376', 'V376A', 'V379', 'V380']].copy()
    
    rec3_data = rec_3.loc[:, ['CASEID', 'V501', 'V502', 'V503', 'V504', 'V505', 'V506', 'V507', 'V508',
                              'V509', 'V510', 'V511', 'V512', 'V513', 'V525', 'V613', 'V714', 'V715']].copy()

    # Get new_var_labels and new_value_labels
    new_keys1 = rec1_data.columns.values
    new_value_labels1 = {key: value for (key, value) in value_labels1.items() if key in new_keys1}
    new_var_labels1 = {key: value for (key, value) in var_labels1.items() if key in new_keys1}

    new_keys2 = rec2_data.columns.values
    new_value_labels2 = {key: value for (key, value) in value_labels2.items() if key in new_keys2}
    new_var_labels2 = {key: value for (key, value) in var_labels2.items() if key in new_keys2}

    new_keys3 = rec3_data.columns.values
    new_value_labels3 = {key: value for (key, value) in value_labels3.items() if key in new_keys3}
    new_var_labels3 = {key: value for (key, value) in var_labels3.items() if key in new_keys3}

    # Generate new column "year" and update new_var_labels
    rec1_data.loc[:, "year"] = year
    new_var_labels1.update({"year": "Year of the survey"})

    # Merge rec1_data, rec2_data and rec3_data using CASE ID
    merged_data = pd.merge(rec1_data, rec2_data, on='CASEID', how='outer')
    merged_data = pd.merge(merged_data, rec3_data, on='CASEID', how='outer')

    # Unify all new_var_labels and new_value_labels
    var_labels = {}
    var_labels.update(new_var_labels1)
    var_labels.update(new_var_labels2)
    var_labels.update(new_var_labels3)

    value_labels = {}
    value_labels.update(new_value_labels1)
    value_labels.update(new_value_labels2)
    value_labels.update(new_value_labels3)

    # Add the data and labels to the all_data dictionary
    all_data[f'year_{year}'] = {'data': merged_data, 'var_labels': var_labels, 'value_labels': value_labels}

# Show all_data dictionary
all_data

# Show all_data dictionary in a more presentable way
print(all_data)
```

9. Use `endes_data_2015_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

```python
# Create a new DataFrame with mean values
mean_key_vars = endes_data_2015_2019.groupby(['year', 'V024']).agg(
    mean_total_children=('V201', 'mean'),
    mean_ideal_children=('V613', 'mean'),
    mean_hb_yr_educ=('V715', 'mean'),
    mean_first_marriage=('V511', 'mean')
).reset_index()


mean_key_vars = mean_key_vars.rename(columns={'V201': 'mean_total_children', 'V613': 'mean_ideal_children', 'V715': 'mean_hb_yr_educ', 'V511': 'mean_first_marriage'})

# We show the first variables of the new dataframe
mean_key_vars.head()
```

10. Merge `mean_key_vars` with `endes_data_2015_2019`. Name this object `final_result`. **Hint: Use merge.**

```python
# Merging both df in one object call "final_result"
final_result = endes_data_2015_2019.merge(mean_key_vars, on = [ 'year' ,'V024' ] , how = "left" , validate = "m:m" )

# We show the first variables of the new dataframe
final_result.head()
```
